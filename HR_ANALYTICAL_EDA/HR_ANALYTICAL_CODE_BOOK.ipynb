```python
# reason of termination of different employees
plt.figure(figsize = (10,12))
plt.bar(y.index , y)
plt.xticks(rotation = 70)
plt.xlabel('reason of termination')
plt.ylabel('count of employees')
plt.show()
```

```python
# no of employees terminated at different position
df2 = df.copy()
df2['TermReason'] = df['TermReason'].replace('N/A-StillEmployed',np.NAN)
df2['TermReason'] = df2['TermReason'].replace('Another position',np.NAN)
df2
```

```python
df2.dropna(inplace = True)
```

```python
y = df2.groupby('Position')['TermReason'].count()
```

```python
y
```

```python
plt.figure(figsize = (12,12))
plt.barh(y.index , y)
plt.xlabel('No of employees')
plt.ylabel('Position')
plt.yticks(rotation = 30)
plt.title('employees fired from different position' )
for index, value in enumerate(y):
    plt.text(value+0.5, index, str(value))   # (x, y, text)
plt.show()
```
