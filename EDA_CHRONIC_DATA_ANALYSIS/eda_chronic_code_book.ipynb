```python
df.columns = ['id','age', 'blood_pressure', 'specific_gravity', 'albumin', 'sugar', 'red_blood_cells', 'pus_cell',
              'pus_cell_clumps', 'bacteria', 'blood_glucose_random', 'blood_urea', 'serum_creatinine', 'sodium',
              'potassium', 'haemoglobin', 'packed_cell_volume', 'white_blood_cell_count', 'red_blood_cell_count',
              'hypertension', 'diabetes_mellitus', 'coronary_artery_disease', 'appetite', 'peda_edema',
              'aanemia', 'class']
```

```python
#  first thing i saw was that there is the age column is present as float converting it to into .. rounding off to floor if <.6 and to ciel if >=.6
```

```python
#  before applying first finding the missing value in age columns
df['age'].isnull().sum()
df = df[~df['age'].isnull()]
```

```python
#  applying transformation to age column
def age_transformation(x):
    num = x- int(x)
    if num>0.6 :
        np.ceil(x)
    else :
        return np.floor(x)


df['age'] = df['age'].apply(age_transformation).astype(int)
```

```python
df.dtypes
```

```python
df.isnull().sum()
```

```python
df.drop('id', axis = 1 , inplace = True) # as it is a identity columns and providing much info 
```

```python
# df['packed_cell_volume'] if this columns is showing volume so why its type is object lets find it out
```

```python
df['packed_cell_volume'].unique() # there are some value containing \t? and \t .. lets replace it 
```

```python
df['packed_cell_volume'] = pd.to_numeric(df['packed_cell_volume'], errors = 'coerce') # coerse = ignore the error
```

```python
df['packed_cell_volume'].unique()
```

```python
# as in the white blood cell count there is count so it should be numeric but showing abject lets fix it 
print(df['white_blood_cell_count'].unique(),'\n')
print(df['red_blood_cell_count'].unique())
```

```python
df['white_blood_cell_count'] = pd.to_numeric(df['white_blood_cell_count'], errors = 'coerce') # coerse = ignore the error
df['red_blood_cell_count'] = pd.to_numeric(df['red_blood_cell_count'], errors = 'coerce') # coerse = ignore the error
```

```python
df.dtypes
```

```python
cat_col = [col for col in df.columns if df[col].dtype == object]
num_col = [col for col in df.columns if df[col].dtype != object]
```

```python
cat_col
```

```python
num_col
```

```python
for col in cat_col:
    print(f"for {col} in df unique values are \n{df[col].unique()} \n")
```

```python
df['diabetes_mellitus'].replace(['\tno' , '\tyes'] , ['no','yes'] , inplace = True)
df['diabetes_mellitus'].unique()
```

```python
df['coronary_artery_disease'].replace(['\tno' , '\tyes'] , ['no','yes'] , inplace = True)
df['coronary_artery_disease'].unique()
```

```python
df['class'] = df['class'].map({'ckd\t' : 1 ,
                 'notckd' : 0,
                 'ckd' : 1 })
df['class'].unique()
```

```python
print(df['class'].dtype)
# as of now we converted it to numeric col moving it to num_col list
num_col.append('class')
```

```python
cat_col.remove('class')
for col in cat_col:
    print(f"for {col} in df unique values are \n{df[col].unique()} \n")
```
